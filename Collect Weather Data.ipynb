```python
import pandas as pd
import numpy as np
import requests
```

## Import the secret key necessary to access the weather API
If you do not have this file, consult the Telegram group

Make sure not to commit the key to git, as that would make it public

```python
secret_key = ""

with open('weather_key.txt') as key_file:
    secret_key = key_file.readline(32)
```

## Retrieve the data

```python
helsinki = "/60.1695200,24.9354500,"
time = "1507670281" # 10 October 2017

weather_request = "https://api.darksky.net/forecast/" + secret_key + helsinki + time

response = requests.get(weather_request)
```

# Parse the response

### One example is to retrieve the weather at the exact time in the request

```python
current_weather = pd.DataFrame.from_dict(response.json()['currently'], orient='index')
current_weather
```

### Or you can get the data for each hour on the day of the request

```python
hourly_weather = pd.DataFrame.from_dict(response.json()['hourly']['data'])
hourly_weather
```
